```python
X = df.drop(columns=['Rain'])  # Fitur
y = df['Rain']  # Target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
# Normalization

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from imblearn import over_sampling
X_train_SMOTE,y_train_SMOTE = over_sampling.SMOTE().fit_resample(X_train,y_train)
```

```python
pd.Series(y_train_SMOTE).value_counts()
```

```python
model = LogisticRegression(max_iter=1000,random_state = 42)
model.fit(X_train_SMOTE,y_train_SMOTE)
```

```python
# prediction with data train 
y_train_pred = model.predict(X_train_SMOTE)
print("Evaluation Data Training:")
print(f"Akurasi: {accuracy_score(y_train_SMOTE, y_train_pred):.4f}")
print(f"Precision: {precision_score(y_train_SMOTE, y_train_pred):.4f}")
print(f"Recall: {recall_score(y_train_SMOTE, y_train_pred):.4f}")
print(f"F1-Score: {f1_score(y_train_SMOTE, y_train_pred):.4f}")
print("\nConfusion Matrix (Training):")
print(confusion_matrix(y_train_SMOTE, y_train_pred))
print("\nClassification Report (Training):")
print(classification_report(y_train_SMOTE, y_train_pred))

# Visualisasi Confusion Matrix (Training)
cm_train = confusion_matrix(y_train_SMOTE, y_train_pred)
sns.heatmap(cm_train, annot=True, fmt='d', cmap='Blues', xticklabels=["No Rain", "Rain"], yticklabels=["No Rain", "Rain"])
plt.title("Confusion Matrix - Training Data")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
y_test_pred = model.predict(X_test)

# Kinerja dan analisis tahap Testing
print("\nEvaluation Model with Data Testing:")
print(f"Akurasi: {accuracy_score(y_test, y_test_pred):.4f}")
print(f"Precision: {precision_score(y_test, y_test_pred):.4f}")
print(f"Recall: {recall_score(y_test, y_test_pred):.4f}")
print(f"F1-Score: {f1_score(y_test, y_test_pred):.4f}")
print("\nConfusion Matrix (Testing):")
print(confusion_matrix(y_test, y_test_pred))
print("\nClassification Report (Testing):")
print(classification_report(y_test, y_test_pred))

# Visualisasi Confusion Matrix (Testing)
cm_test = confusion_matrix(y_test, y_test_pred)
sns.heatmap(cm_test, annot=True, fmt='d', cmap='Blues', xticklabels=["No Rain", "Rain"], yticklabels=["No Rain", "Rain"])
plt.title("Confusion Matrix - Testing Data")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
# tunning treshold
y_probs = model.predict_proba(X_test)[:,1]

precision,recall,thresholds = precision_recall_curve(y_test,y_probs)
f1_scores = 2 * (precision * recall) / (precision + recall)
optimal_idx = f1_scores.argmax()
optimal_threshold = thresholds[optimal_idx]

print(f"Optimal Threshold: {optimal_threshold:.2f}")
print(f"Precision Optimal Threshold: {precision[optimal_idx]:.2f}")
print(f"Recall Optimal Threshold: {recall[optimal_idx]:.2f}")
print(f"F1-Score Optimal Threshold: {f1_scores[optimal_idx]:.2f}")
```

```python
# ROC-AUC
roc_auc = roc_auc_score(y_test,y_probs)
print(f"ROC-AUC data tes : {roc_auc : .2f}")

y_probs_train = model.predict_proba(X_train_SMOTE)[:,1]
roc_auc_train = roc_auc_score(y_train_SMOTE,y_probs_train)
print(f"ROC-AUC data train : {roc_auc_train : .2f}")
```

```python
fpr_test, tpr_test, _ = roc_curve(y_test, y_probs)
fpr_train, tpr_train, _ = roc_curve(y_train_SMOTE, y_probs_train)

# Plot ROC Curve
plt.figure(figsize=(10, 6))

# Plot ROC Curve untuk data test
plt.plot(fpr_test, tpr_test, color='blue', lw=2, label=f'Test ROC curve (AUC = {roc_auc:.2f})')

# Plot ROC Curve untuk data train
plt.plot(fpr_train, tpr_train, color='red', lw=2, label=f'Train ROC curve (AUC = {roc_auc_train:.2f})')

# Garis diagonal (Random Classifier)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')

# Menambahkan label dan judul
plt.title('ROC Curve')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.legend(loc='lower right')

# Menampilkan plot
plt.show()
```

```python
y_pred_optimal = (y_probs >= optimal_threshold).astype(int)

# Evaluasi metrik dengan threshold optimal
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_optimal))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_optimal))
```
